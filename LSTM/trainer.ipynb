```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import torch.nn.functional as F
from torch.utils.data import DataLoader , Dataset
import numpy as np
import matplotlib.pyplot as plt

import os
from dotenv import load_dotenv
import requests
import json
```

```python
alphavantage_api_key = os.getenv("ALPHAVANTAGE_API_KEY")
alphavantage_api_url = f"https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=MSFT&apikey={alphavantage_api_key}"
```

```python
data = requests.get(alphavantage_api_url)
data = json.loads(data.text)
time = list(data['Time Series (Daily)'].keys())


#draw the data
open_price = []
close_price = []
high_price = []
low_price = []

for i in time:
    open_price.append(float(data['Time Series (Daily)'][i]['1. open']))
    close_price.append(float(data['Time Series (Daily)'][i]['4. close']))
    high_price.append(float(data['Time Series (Daily)'][i]['2. high']))
    low_price.append(float(data['Time Series (Daily)'][i]['3. low']))

plt.plot(open_price)
plt.plot(close_price)
plt.show()
```
